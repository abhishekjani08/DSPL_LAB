```python
import pandas as pd
import numpy as np
from pyspark.sql import SparkSession
import pyspark.sql.functions as F
```

```python
# create a SparkSession
spark = SparkSession.builder.appName("House Rent").getOrCreate()
```

```python
df1 = spark.read.csv("D:\sem-6\DSPL\DSPL_LAB-main\House_Rent_main6-main1.csv",header=True, inferSchema=True)
```

```python
df = df1
df.printSchema() # print the schema of the DataFrame
```

```python
# visualize the data
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(6,5))
sns.histplot(df.toPandas(), bins=5)
```

```python
df.head()
```

```python
# visualize the data
plt.figure(figsize=(8,6))
sns.boxplot(data=df.toPandas(), x="City", y="BHK")
plt.show()
```

```python
# stop the SparkSession
spark.stop()
```

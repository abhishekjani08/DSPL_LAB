```python
#import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('House_Rent_main2.csv')
df1=df
df1.sample(5)
```

```python
import warnings
warnings.filterwarnings('ignore')
plt.figure(figsize=(16,5))
plt.subplot(1,2,1)
sns.distplot(df1['BHK'])
plt.subplot(1,2,2)
sns.distplot(df1['Bathroom'])
plt.show()
```

```python
#Checking outliers using boxplot
print(plt.figure(figsize = (10, 4), dpi = 100)) 
sns.boxplot(x = "BHK", data = df1) 

print(plt.figure(figsize = (10, 4), dpi = 100)) 
sns.boxplot(x = "Bathroom", data = df1)
```

```python
sns.boxplot(x ='Furnishing Status', y ='Bathroom', data = df1, hue ='BHK')
```

```python
# compare plots before trimming
plt.figure(figsize=(16,8))
plt.subplot(2,2,1)
sns.distplot(df1['BHK'])
plt.subplot(2,2,2)
sns.boxplot(df1['BHK'])
plt.show()
```

```python
# finding boundary values for bmi using z-score
print("Highest allowed",df1['BHK'].mean() + 3*df1['BHK'].std())
print("Lowest allowed",df1['BHK'].mean() - 3*df1['BHK'].std())
```

```python
# finding outliers for bmi
df1[(df1['BHK'] > 4.580627788100002) | (df1['BHK'] < -0.4129076026807015)]
```

```python
# Trimming of outliers for bmi
df2 = df1[(df1['BHK'] < 4.580627788100002) & (df1['BHK'] > -0.4129076026807015)]
df2
```

```python
# compare plots after trimming
plt.figure(figsize=(16,8))
plt.subplot(2,2,1)
sns.distplot(df2['BHK'])
plt.subplot(2,2,2)
sns.boxplot(df2['BHK'])
plt.show()
```
